```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("G:\diamond data\diamonds.csv")
df
```

```python
df.describe(include='all')
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
num_cols
```

```python
cat_cols  =df.select_dtypes(include=['object']).columns
cat_cols
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower,upper)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=['price'],axis=1)
X
```

```python
y=df['price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=45)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_num_cols = X.select_dtypes(include=["int64", "float64"]).columns
X_cat_cols = X.select_dtypes(include=["object", "category"]).columns
print(num_cols)
print(cat_cols)
```

```python
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
        ('scaler',StandardScaler())
    ]
)

cat_pipeline = Pipeline([
        ('scaler',OrdinalEncoder())
        
    ]
)
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num',num_pipeline,X_num_cols),
        ('cat',cat_pipeline,X_cat_cols)
    ]
)

X_transform_df = preprocessor.fit_transform(X)
X_transform_df = pd.DataFrame(X_transform_df,columns=X.columns)
X_transform_df
```

```python
from sklearn.feature_selection import mutual_info_regression
mi_scores = mutual_info_regression(
    X_transform_df,
    y,
    random_state=42
)

mi = pd.Series(
    mi_scores,
    index=preprocessor.get_feature_names_out()
).sort_values(ascending=False)

print(mi)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
```

```python
from sklearn.pipeline import make_pipeline
final_pipeline = Pipeline([
    ('preprocessing',preprocessor),
    ('model',model)
])

final_pipeline.fit(X_train,y_train)
```

```python
y_pred = final_pipeline.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
print('testing data: ')
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R²:", r2_score(y_test, y_pred))
```

```python
y_train_pred = final_pipeline.predict(X_train)
y_train_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
print('training data: ')
print("MAE:", mean_absolute_error(y_train, y_train_pred))
print("MSE:", mean_squared_error(y_train, y_train_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("R²:", r2_score(y_train, y_train_pred))
```
